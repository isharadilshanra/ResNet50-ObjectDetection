pairwise squared Euclidean distances

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# matrix dimensions
m = 10 # for x
n = 10 # for y
d = 10 # dataset size

# genrate random data
x = np.random.rand(m, d)
y = np.random.rand(n, d)

# pairwise squared Euclidean distances
# need an m × n matrix Z with Z(i , j) =∥xi −yj ∥2

# Zi,j​=∥xi​−yj​∥2=(xi​−yj​)T(xi​−yj​)=∥xi​∥2+∥yj​∥2−2xiT​yj

# Compute the squared norms of each row in X and Y
X = np.sum(x**2, axis=1).reshape(-1, 1)  # Shape (m, 1)

# need the transpose of y, reshape(1,-1)
Y = np.sum(y**2, axis=1).reshape(1, -1)  # Shape (1, n)

# Compute the pairwise squared distances using broadcasting
Z = X + Y - 2 * np.dot(x, y.T)

print(Z)

```

```python
#  method 2
# Reshape X to (m, 1, d) and Y to (1, n, d)
X_expanded = x[:, np.newaxis, :]  # Shape: (m, 1, d)
Y_expanded = y[np.newaxis, :, :]  # Shape: (1, n, d)

# Subtract, square, and sum over the last axis
distances = np.sum((X_expanded - Y_expanded) ** 2, axis=2)  # Shape: (m, n)

distances
```

**Explanation of Steps:**

    Reshape X to (m, 1, d): This reshaping is crucial because it allows NumPy to broadcast X against Y properly.
    Now, every row of X will be subtracted from every row of Y.

    Reshape Y to (1, n, d): Similarly, reshaping Y ensures that Y is broadcasted to match X.
    
    Now, every row of Y can be subtracted from every row of X.

    Broadcast and Subtract: Broadcasting allows every row in X to be subtracted from every row in Y, giving a matrix of differences with shape (m, n, d).

    Square and Sum: After computing the differences, we square the result element-wise and sum along the last axis (axis=2), which corresponds to the feature dimension d.
    
    This results in a matrix of shape (m, n) that contains the squared Euclidean distances between each pair of rows from X and Y.

**Transfer lerning using resnet50 (tensor flow/ caltech101 dataset)**

```python
# import libtraries
import os
import datetime
import tensorflow as tf
import numpy as np
import tensorflow_datasets as tfds
from matplotlib import pyplot as plt
from tensorflow.keras import layers
```

**Step 1: Load the Caltech-101 Dataset**

```python

# Load the dataset from TensorFlow Datasets
(dataset_train, dataset_test), dataset_info = tfds.load(
    'caltech101',
    split=['train', 'test'],
    with_info=True,
    as_supervised=True  # This will return (image, label) pairs
)

# Preprocess the dataset to be used with ResNet-50
def preprocess_image(image, label):
    image = tf.image.resize(image, (224, 224))  # ResNet-50 requires input size (224, 224)
    #image = image / 255.0  # Normalize to [0, 1] range
    return image, label

# Apply preprocessing
dataset_train = dataset_train.map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)
dataset_test = dataset_test.map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)
```

**Step 2: Load Pre-trained ResNet-50 and Remove the Last Layer**

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model

# Load the pre-trained ResNet-50 model without the top (classification) layer
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Add a global average pooling layer to get embeddings
x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)

# Create a new model that outputs the embeddings
model = Model(inputs=base_model.input, outputs=x)

# Freeze the base model layers to prevent them from training
for layer in base_model.layers:
    layer.trainable = False

model.summary()
```

**Step 3: Extract Embeddings for the Dataset**

```python

# Function to extract embeddings from the dataset
def extract_embeddings(model, dataset):
    embeddings = []
    labels = []
    for images, label_batch in dataset:
        emb = model.predict(images)
        embeddings.append(emb)
        labels.append(label_batch)
    return np.concatenate(embeddings), np.concatenate(labels)

# Extract embeddings for train and test sets
train_embeddings, train_labels = extract_embeddings(model, dataset_train)
test_embeddings, test_labels = extract_embeddings(model, dataset_test)
```

**Step 4: Perform k-NN Classification**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Create a k-NN classifier
knn = KNeighborsClassifier(n_neighbors=5)

# Train the k-NN classifier on the train embeddings
knn.fit(train_embeddings, train_labels)

# Predict the labels on the test set
test_predictions = knn.predict(test_embeddings)

# Calculate accuracy
accuracy = accuracy_score(test_labels, test_predictions)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```

**c. Fine-tune an image classification network end-to-end**

```python
base_model = tf.keras.applications.ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True  # Fine-tune the pre-trained layers

model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(102, activation='softmax')  # 101 object classes
])

model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2)
])

def augment(image, label):
    image = data_augmentation(image)
    return image, label

dataset_train = dataset_train.map(augment).batch(32).prefetch(tf.data.AUTOTUNE)

history = model.fit(dataset_train, validation_data=dataset_test, epochs=10)

test_loss, test_accuracy = model.evaluate(dataset_test)
print(f'Fine-tuned Model Test Accuracy: {test_accuracy * 100:.2f}%')
```

```python
import os
import datetime
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

# Load the Caltech101 dataset from TensorFlow Datasets
import tensorflow_datasets as tfds
train_dataset, test_dataset = tfds.load('caltech101', split=['train', 'test'], as_supervised=True)

# Define the data augmentation pipeline
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2)
])

# Define a function to preprocess the images for ResNet50 and ensure correct image size
def preprocess(image, label):
    # Resize to 224x224 as expected by ResNet50 and normalize using preprocess_input
    image = tf.image.resize(image, [224, 224])
    image = tf.keras.applications.resnet50.preprocess_input(image)
    return image, label

# Define a function to apply data augmentation
def augment(image, label):
    image = data_augmentation(image)
    return image, label

# Apply preprocessing and augmentation to the training data
train_dataset = train_dataset.map(preprocess).map(augment).batch(32).prefetch(tf.data.AUTOTUNE)

# Apply only preprocessing to the test data
test_dataset = test_dataset.map(preprocess).batch(32).prefetch(tf.data.AUTOTUNE)

# Load the pre-trained ResNet50 model (with weights from ImageNet, excluding the top layers)
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Allow fine-tuning of the pre-trained layers
base_model.trainable = True

# Define the full model
model = tf.keras.Sequential([
    base_model,
    GlobalAveragePooling2D(),  # Global pooling to reduce the dimensions
    tf.keras.layers.Dense(102, activation='softmax')  # Output layer for 101 object classes
])

# Compile the model with an appropriate optimizer and loss function
model.compile(optimizer=Adam(1e-4),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# load tensor board
logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = tf.keras.callbacks.TensorBoard(logdir, histogram_freq=1)

# Train the model for 120 epochs
history = model.fit(train_dataset, validation_data=test_dataset, epochs=120, callbacks=[tensorboard_callback])

# Evaluate the model on the test dataset
test_loss, test_accuracy = model.evaluate(test_dataset)
print(f'Fine-tuned Model Test Accuracy: {test_accuracy * 100:.2f}%')
print(f'Fine-tuned Model Test Lost: {test_loss * 100:.2f}%')
```
